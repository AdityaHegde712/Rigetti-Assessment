```python
import os
from glob import glob

TRAIN_DATA = "../dataset/train"
TEST_DATA = "../dataset/test"
```

```python
train_classes = glob(f"{TRAIN_DATA}/*")
test_classes = glob(f"{TEST_DATA}/*")
print(*train_classes, sep="\n", end="\n\n")
print(*test_classes, sep="\n")
```

```python
# %%
from collections import Counter

def inspect_image_sizes(class_dirs, max_images_per_class=None):
    size_counter = Counter()

    for cls in class_dirs:
        images = os.listdir(cls)
        if max_images_per_class:
            images = images[:max_images_per_class]

        for img_name in images:
            img_path = os.path.join(cls, img_name)
            try:
                with Image.open(img_path) as img:
                    size_counter[img.size] += 1  # (width, height)
            except Exception as e:
                print(f"Error loading {img_path}: {e}")

    return size_counter

# Inspect (limit optional for speed, e.g., 100)
train_sizes = inspect_image_sizes(train_classes, max_images_per_class=50)

# Print summary
print("Unique image sizes:", len(train_sizes))
print("\nMost common sizes:")
for size, count in train_sizes.most_common(10):
    print(f"{size}: {count}")

# Optional: check if all images are same size
if len(train_sizes) == 1:
    print("\nAll images have the same size.")
else:
    print("\nImages have varying sizes.")
```
